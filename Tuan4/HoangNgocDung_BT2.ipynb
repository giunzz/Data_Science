```python
import numpy as np
import pandas as pd
```

```python
file_path = "clean_feature.csv"
df = pd.read_csv(file_path)
df_clean = df.dropna().drop_duplicates()
df_clean
```

1> Lập hirachical index theo các level sau : Năm, ngày trong tuần, giờ, dịch vụ

2> thay đổi level theo ngày, năm, dịch vụ

3> Tính phân phối các dịch vụ theo ngày

4> tính phân phối các dịch vụ theo giờ

```python
df_clean["Day"] = pd.to_datetime(df_clean["Day"])
df_clean["Year"] = df_clean["Day"].dt.year
# print(df_clean["Year"])
df_hier = df_clean.set_index(["Year", "DoWeek", "hour", "ServiceID"])
df_hier
# print(df_hier.head(20))
```

```python
df_hier = df_hier.swaplevel('DoWeek', 'Year')
df_hier = df_hier.swaplevel('ServiceID', 'hour')
df_hier
```

```python
# Tính phân phối theo ngày
pivot_counts_day = pd.pivot_table(
    df_hier,
    index="DoWeek",
    columns="ServiceID",
    values="count",  
    aggfunc="sum",   
    fill_value=0
)
pd.pivot_table(
    df_hier,
    index="DoWeek",
    columns="ServiceID",
    values="count",  
    aggfunc="sum",   
    fill_value=0
)

```

```python
print("Ngày cao điểm:")
result = pd.DataFrame({
    "day_max": pivot_counts_day.idxmax(),
    "Value_max": pivot_counts_day.max()
})
result
```

```python
# Tính phân phối theo giờ
pivot_counts_hour = pd.pivot_table(
    df_hier,
    index="hour",
    columns="ServiceID",
    values="count",  
    aggfunc="sum",   
    fill_value=0
)
pd.pivot_table(
    df_hier,
    index="hour",
    columns="ServiceID",
    values="count",  
    aggfunc="sum",   
    fill_value=0
)

```

```python
print("Giờ cao điểm:")
result = pd.DataFrame({
    "Hour_max": pivot_counts_hour.idxmax(),
    "Value_max": pivot_counts_hour.max()
})

result
```
