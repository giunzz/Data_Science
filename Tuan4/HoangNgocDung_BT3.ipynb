```python
df = pd.read_csv("clean_feature.csv")
df_clean = df.dropna().drop_duplicates() # Xoá các dòng bị thiếu và trùng lặp
df_clean
```

```python
# Cho thời điểm t (tính theo đơn vị giờ) , hãy querry dataframe nằm trong khoảng thời gian t-T. Trong đó T là giá trị biết trước tính bằng giờ
import pandas as pd

df["datetime"] = pd.to_datetime(df["Day"]) + pd.to_timedelta(df["hour"], unit="h")

def query_interval(df, t_datetime, T):
    start = t_datetime - pd.Timedelta(hours=T)
    return df[(df["datetime"] >= start) & (df["datetime"] <= t_datetime)]

t = pd.to_datetime("2021-12-30 3:00")
result = query_interval(df, t, 5)
result.sort_index()
# print(result[["Day", "hour", "datetime"]])
```

```python
# Dùng groupby để nhóm các dữ liệu của cùng một dịch vụ thành từng nhóm.
grouped = df.groupby("ServiceID")
grouped.describe()
```

```python
# Rút trích các đặc trưng
agg_df = df.groupby("ServiceID").agg(
    total_calls=("count", "sum"),          # a> Tổng số lần gọi
    total_success=("passed", "sum"),       # b> Tổng số lần gọi thành công
    mean_exec_time=("period", "mean"),     # c> Trị trung bình thời gian thực thi
    mean_data_usage=("data", "mean")       # d> Trị trung bình dữ liệu sử dụng
).reset_index()

agg_df
```

```python
# Gọi M là số thời điểm (t) có thể lấy được trong bộ dữ liệu, và N là số dịch vụ, hãy thành lập một ma trận kích thước M*(4*N) chứa các đặc trưng của câu 5 cho mỗi dịch vụ


features = df.groupby(["datetime", "ServiceID"]).agg(
    total_calls=("count", "sum"), # a> Tổng số lần gọi
    total_success=("passed", "sum"), # b> Tổng số lần gọi thành công
    mean_exec_time=("period", "mean"), # c> Trị trung bình thời gian thực thi
    mean_data_usage=("data", "mean") # d> Trị trung bình dữ liệu sử dụng
)

matrix = features.unstack(level="ServiceID")

matrix = matrix.sort_index(axis=1, level=1)
matrix
```
