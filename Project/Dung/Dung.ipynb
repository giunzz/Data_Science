```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

element_list = []

# Set up Chrome options (optional)
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode (optional)
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Use a proper Service object
service = Service(ChromeDriverManager().install())

for page in range(1, 3):
    # Initialize driver properly
    driver = webdriver.Chrome(service=service, options=options)

    # Load the URL
    url = f"https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page=%7Bpage%7D"
    driver.get(url)
    time.sleep(2)  # Optional wait to ensure page loads

    # Extract product details
    titles = driver.find_elements(By.CLASS_NAME, "title")
    prices = driver.find_elements(By.CLASS_NAME, "price")
    descriptions = driver.find_elements(By.CLASS_NAME, "description")
    ratings = driver.find_elements(By.CLASS_NAME, "ratings")

    # Store results in a list
    for i in range(len(titles)):
        element_list.append([
            titles[i].text,
            prices[i].text,
            descriptions[i].text,
            ratings[i].text
        ])

    driver.quit()

# Display extracted data
for row in element_list:
    print(row)
```

```python
from playwright.sync_api import sync_playwright


def main():
    with sync_playwright() as p:
        browser = p.chromium.launch(headless=False)
        page = browser.new_page()
        page.goto('https://quotes.toscrape.com/')
        all_quotes = page.query_selector_all('.quote')

        for quote in all_quotes:
            text = quote.query_selector('.text').inner_text()
            author = quote.query_selector('.author').inner_text()
            print({'Author': author, 'Quote': text})

        page.wait_for_timeout(10000)
        browser.close()


if __name__ == '__main__':
    main()
    
```

```python
import re, csv, time
from urllib.parse import urljoin
from playwright.sync_api import sync_playwright

BASE = "https://viasm.edu.vn/hoat-dong-khoa-hoc/hoat-dong-trong-nam"

def scrape_viasm(output_csv="viasm_hoat_dong.csv", year_text=None, max_pages=None):
    rows = []

    with sync_playwright() as p:
        browser = p.chromium.launch(headless=True)
        # Tạo 2 tab: 1 tab duyệt danh sách, 1 tab mở chi tiết
        context = browser.new_context()
        list_page = context.new_page()
        detail_page = context.new_page()

        list_page.goto(BASE, wait_until="networkidle")

        # (Tuỳ chọn) Lọc theo năm nếu muốn, ví dụ year_text="Năm 2025"
        if year_text:
            year_locator = list_page.get_by_text(year_text, exact=True)
            if year_locator.count():
                year_locator.first.click()
                list_page.wait_for_load_state("networkidle")

        page_idx = 1
        while True:
            # Lấy tất cả link sự kiện ở trang hiện tại (href bắt đầu bằng /hdkh/)
            anchors = list_page.locator('a[href^="/hdkh/"]').all()
            hrefs = []
            seen = set()
            for a in anchors:
                href = a.get_attribute("href")
                text = (a.text_content() or "").strip()
                if not href or href in seen or not text:
                    continue
                seen.add(href)
                hrefs.append((urljoin(BASE, href), text))

            # Mở từng link chi tiết để lấy thông tin có cấu trúc
            for url, fallback_title in hrefs:
                detail_page.goto(url, wait_until="networkidle")

                # Tiêu đề
                title = ""
                for sel in ["h1", "h2", "h3"]:
                    if detail_page.locator(sel).count():
                        title = (detail_page.locator(sel).first.text_content() or "").strip()
                        if title:
                            break
                if not title:
                    title = fallback_title

                # Lấy toàn bộ text rồi regex các trường
                body = detail_page.locator("body").inner_text()

                def pick(label):
                    m = re.search(rf"{label}\s*:\s*(.+)", body)
                    return m.group(1).strip() if m else ""

                time_range = pick("Thời gian")
                location   = pick("Địa điểm")
                speaker    = pick("Báo cáo viên")

                rows.append({
                    "title": title,
                    "time": time_range,
                    "location": location,
                    "speaker": speaker,
                    "url": url,
                })

                time.sleep(0.15)  # lịch sự 1 chút

            # Chuyển trang: tìm nút "Sau"
            next_btn = list_page.locator('a:has-text("Sau")')
            if not next_btn.count() or not next_btn.first.is_visible():
                break

            # Dừng sớm nếu đặt giới hạn số trang
            if max_pages and page_idx >= max_pages:
                break

            next_btn.first.click()
            list_page.wait_for_load_state("networkidle")
            page_idx += 1

        browser.close()

    # Lưu CSV
    with open(output_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["title","time","location","speaker","url"])
        writer.writeheader()
        writer.writerows(rows)

    print(f"Đã lưu {len(rows)} bản ghi vào {output_csv}")

if __name__ == "__main__":
    scrape_viasm(year_text="Năm 2025", max_pages=3)
    # scrape_viasm()
```
