### Thông tin chung:



```python
import pandas as pd
file_path = "clean_feature.csv"
df = pd.read_csv(file_path)
```

1> Có bao nhiêu dịch vụ

2> Mỗi dịch vụ được triển khai lần đầu vào thời gian nào

3> thời gian chạy trung bình của mỗi dịch vụ

4> Dung lượng chạy trung bình của mỗi dịch vụ

```python
num_services = df["ServiceID"].nunique()
first_deploy = df.groupby("ServiceID")["Day"].min()
print(f"Tổng số dịch vụ: {num_services}")
print("Ngày triển khai đầu tiên của từng dịch vụ:")
print(first_deploy)
```

```python
avg_runtime = df.groupby("ServiceID")["period"].mean()
print("Thời gian chạy trung bình của từng dịch vụ:")
print(avg_runtime)
```

```python
avg_data = df.groupby("ServiceID")["data"].mean()
print("Dung lượng chạy trung bình của từng dịch vụ:")
print(avg_data)
```

## Làm sạch dữ liệu:

1> Với dữ liệu đã cho, tìm xem có dữ liệu nào bị NaN, bị duplicate ko? Hãy xử lý các dữ liệu đó.

```python
print("Số lượng NaN trong từng cột:", df.isna().sum())
print("Số dòng duplicate:", df.duplicated().sum())
df_clean = df.dropna()
df_clean = df_clean.drop_duplicates()
```

```python
print("Sau khi làm sạch:")
print("NaN:", df_clean.isna().sum().sum())
print("Duplicate:", df_clean.duplicated().sum())
print("Kích thước dữ liệu:", df_clean.shape)

df_clean.to_csv("cleaned_feature.csv", index=False)
```

2> cho một thời điểm T, hãy viết chương trình truy vấn tất cả các sự kiện xảy trong trong khoảng thời gian T-2 tới T.


```python
def query2(df, T):
    T = int(2)
    start_time = T - 2
    mask = (df["hour"] > start_time) & (df["hour"] <= T)
    return df.loc[mask]

df_clean = pd.read_csv("cleaned_feature.csv")
T = "12"
events = query2(df_clean, T)
```

3> Từ những dữ kiện truy vấn đó trả lời câu hỏi sau:

a> Có bao nhiêu dịch vụ được gọi trong khoảng thời gian đó,

b> Thời gian chạy trung bình của từng dịch vụ là bao nhiêu,

c> dung lượng chạy của tường dịch vụ theo thời gian là bao nhiêu

```python
print(f"Số sự kiện trong khoảng (T-2h, T] với T = {T}: {len(events)}")

num_services = events["ServiceID"].nunique()
avg_runtime = events.groupby("ServiceID")["period"].mean()
total_data = events.groupby("ServiceID")["data"].sum()
```

```python
print(f"Số dịch vụ được gọi: {num_services}")
print(f"Thời gian chạy trung bình theo từng dịch vụ:\n{avg_runtime}")
print(f"Tổng dung lượng chạy theo từng dịch vụ:\n{total_data}")
```
