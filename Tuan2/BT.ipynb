3. Dictionary
Táº¡o dictionary students = {"An": 8, "BÃ¬nh": 9, "Chi": 7} (key lÃ  tÃªn, value lÃ  Ä‘iá»ƒm).
ThÃªm má»™t sinh viÃªn má»›i "Dung": 10.
Cáº­p nháº­t Ä‘iá»ƒm cá»§a "An" thÃ nh 9.
In ra danh sÃ¡ch táº¥t cáº£ sinh viÃªn cÃ³ Ä‘iá»ƒm >= 9.

```python
students = {"An": 8, "BÃ¬nh": 9, "Chi": 7}
students["Dung"] = 10
students["An"] = 9
for name, score in students.items():   # .items() tráº£ vá» (key, value)
    if score >= 9:
        print(name, score)
```

4. Set
Táº¡o hai set:
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
TÃ¬m giao, há»£p, vÃ  hiá»‡u cá»§a hai set trÃªn.
In káº¿t quáº£.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Giao (intersection)
giao = A & B
# Há»£p (union)
hop = A | B
# Hiá»‡u (difference: A trá»« B)
hieu = A - B

print("Giao:", giao)
print("Há»£p:", hop)
print("Hiá»‡u (A - B):", hieu)
```

4. Biáº¿n Ä‘á»•i chuá»—i trong list
Cho list words = ["apple", "banana", "cherry"].

Táº¡o list upper_words chá»©a táº¥t cáº£ cÃ¡c tá»« viáº¿t hoa.
Táº¡o list lengths chá»©a Ä‘á»™ dÃ i cá»§a tá»«ng tá»«.

5. Lá»“ng for (nested for)
Cho hai list:

list1 = [1, 2, 3]

list2 = ['a', 'b', 'c']

Táº¡o list má»›i pairs chá»©a táº¥t cáº£ cáº·p (x, y) vá»›i x tá»« list1, y tá»« list2.
ðŸ‘‰ Gá»£i Ã½: dÃ¹ng nested for hoáº·c list comprehension lá»“ng nhau.

```python
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']

pairs = [(x, y) for x in list1 for y in list2]
print(pairs)
```

6. Thá»±c hÃ nh nÃ¢ng cao
Cho list matrix = [[1,2,3],[4,5,6],[7,8,9]].
DÃ¹ng list comprehension Ä‘á»ƒ láº¥y táº¥t cáº£ pháº§n tá»­ ra thÃ nh 1 list pháº³ng:

[1,2,3,4,5,6,7,8,9]

DÃ¹ng list comprehension Ä‘á»ƒ láº¥y cá»™t Ä‘áº§u tiÃªn cá»§a ma tráº­n: [1,4,7].

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flattened = [num for row in matrix for num in row]
print(flattened)
first_collum = [row[0] for row in matrix]
first_collum
```

## BÃ i táº­p Python: for + enumerate + list

1. In chá»‰ sá»‘ vÃ  giÃ¡ trá»‹
Cho list fruits = ["apple", "banana", "cherry", "date"].

In ra dáº¡ng:

```python
fruits = ["apple", "banana", "cherry", "date"]
for i in range(len(fruits)):
    print(i , " ",fruits[i])
```

2. TÃ¬m vá»‹ trÃ­ pháº§n tá»­ thá»a mÃ£n Ä‘iá»u kiá»‡n
Cho list numbers = [10, 25, 30, 45, 50].

In ra vá»‹ trÃ­ (index) cá»§a táº¥t cáº£ pháº§n tá»­ chia háº¿t cho 5.
Output mong muá»‘n:

Pháº§n tá»­ 10 á»Ÿ vá»‹ trÃ­ 0

```python
numbers = [10, 25, 30, 45, 50]
for i in range(len(numbers)):
    if numbers[i] % 5 == 0:
        print(f"Phan tu {numbers[i]} o vi tri {i}")
```

3. Thay Ä‘á»•i pháº§n tá»­ dá»±a theo vá»‹ trÃ­
Cho list words = ["a", "b", "c", "d"].

DÃ¹ng enumerate Ä‘á»ƒ biáº¿n thÃ nh list má»›i new_words, trong Ä‘Ã³ má»—i pháº§n tá»­ lÃ  chá»‰ sá»‘ + kÃ½ tá»±, vÃ­ dá»¥:

```python
words = ["a", "b", "c", "d"]
new_words = [f"{i}-{w}" for i, w in enumerate(words)]
print(new_words)
```

5. Lá»“ng enumerate vá»›i list comprehension
